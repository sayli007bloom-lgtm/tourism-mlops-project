```python
#Load train/test from Hugging Face;
import pandas as pd

# Load train-test data (local files you already created)
X_train = pd.read_csv("tourism_project/data/X_train.csv")
X_test  = pd.read_csv("tourism_project/data/X_test.csv")
y_train = pd.read_csv("tourism_project/data/y_train.csv")
y_test  = pd.read_csv("tourism_project/data/y_test.csv")

# Convert y to proper format
y_train = y_train.squeeze()
y_test = y_test.squeeze()

print("✅ Data loaded successfully")
```

```python
#Build model;
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
#Tune model;
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10, 15]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring='accuracy'
)

grid.fit(X_train, y_train)

best_model = grid.best_estimator_

print("Best Params:", grid.best_params_)
```

```python
#Evaluate model;
from sklearn.metrics import accuracy_score, classification_report

y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
